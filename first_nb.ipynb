## Retrieving data from all pdfs

```python
# importing required modules
bodies = []
names = ['dali','warhol','banksy','marcchagall','picasso','rembrandt','kaws','leoneardtsuguharufoujita','yayoi-kusama','zhangdaqian']
# creating a pdf file object
for name in names:
    open_name = './' + name + '.pdf'
    pdfFileObj = open(open_name, 'rb')
    # creating a pdf reader object
    pdfReader = PyPDF2.PdfReader(pdfFileObj)
    # printing number of pages in pdf file
    num_pages = len(pdfReader.pages)
    print(f"Number of pages: {num_pages}")
    # loop through all pages and extract text
    for page_num in range(num_pages):
        pageObj = pdfReader.pages[page_num]
        text = pageObj.extract_text()
        bodies.append(text)
        print(f"Page {page_num+1}:\n{name}")
        # validate the process body of text is correct format
        if type(text) is str:
            print('Successfully pulled text from page.\n')
        else:
            print('Failed to pull text.')
    # closing the pdf file object
    pdfFileObj.close()
    print(f'+++++++++++++++++++++\nFinished text from {name} pdf\n+++++++++++++++++++++\n')
```

```python
# function for isolating and pulling 
def remove_ch_jp_characters(text):
  """Removes all chinese characters from a string."""
  pattern = re.compile(r'[\u3040-\u30ff\u3400-\u4dbf\u4e00-\u9fff\uf900-\ufaff\uff66-\uff9f]')
  return pattern.sub('', text)
```

```python
import re
import pandas as pd

# Define the text
text = '''
<Salvador Dali>\n<Les Amours Jaunes (Michel et\nLopsinger 693-702; Field 74-15)>\n<the complete set of ten drypoint etchings with\ngilding>\n<Height 39.8 x Width 30 x Depth 2.4 cm>.\n<Height 15.669 x Width 11.811 x Depth 0.945 in.>\n<1974><13 December 2022><11 months ago>\n<Bonhams London>\n<Prints and Multiples - [Lot 00025]>\n<est. 5,000 - 7,000 GBP> <9,562GBP>\n<est. 6,187 - 8,662 USD> <11,832USD>\n| 37% est
'''

# Define individual regex patterns
patterns = [
    r'<([^>]+)>',
    r'est\. (\d{1,3}(?:,\d{3})*)(?:\.\d+)? - (\d{1,3}(?:,\d{3})*)(?:\.\d+)? GBP',
    r'est\. (\d{1,3}(?:,\d{3})*)(?:\.\d+)? - (\d{1,3}(?:,\d{3})*)(?:\.\d+)? USD',
    r'(\d+)% est'
]

# Initialize an empty list to store matches
matches = []

# Apply each regex pattern to the text
for pattern in patterns:
    match = re.findall(pattern, text)
    if match:
        matches.extend(match)
    else:
        matches.extend([None, None])

# Define column names
columns = [
    'artist', 'piece_name', 'medium', 'size_cm', 'size_in',
    'date_created', 'date_sold', 'time_ago', 'auction_house',
    'lot', 'estimated_price_gbp_low', 'estimated_price_gbp_high',
    'estimated_price_usd_low', 'estimated_price_usd_high', 'percentage_est', 'x', 'y']


# Create a DataFrame with the matches
df = pd.DataFrame([matches], columns=columns)

# Print the DataFrame
df
```
